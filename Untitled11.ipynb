```python
import cv2
import numpy as np
import zipfile
import os
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from google.colab.patches import cv2_imshow
from imutils import contours
import subprocess # Added for robust unzip
from sklearn.model_selection import train_test_split # Added for accuracy metric

# -------------------------------
# STEP 1: LOAD IMAGE
# -------------------------------
image = cv2.imread("CAR.jpeg")
orig = image.copy()

# -------------------------------
# STEP 2: PREPROCESSING
# -------------------------------
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Gaussian Filter
gray = cv2.GaussianBlur(gray, (5, 5), 0)

# Binarization (Otsu)
_, binary = cv2.threshold(gray, 0, 255,
                          cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# -------------------------------
# STEP 3: EDGE DETECTION
# -------------------------------
edges = cv2.Canny(gray, 30, 90)

# -------------------------------
# STEP 4: CONTOUR DETECTION
# -------------------------------
contours_cv, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours_cv = sorted(contours_cv, key=cv2.contourArea, reverse=True)[:20]

plate = None

for cnt in contours_cv:
    peri = cv2.arcLength(cnt, True)
    approx = cv2.approxPolyDP(cnt, 0.018 * peri, True)

    if len(approx) == 4:
        plate = approx
        break

if plate is None:
    print("Plate not detected")
    exit()

# Draw detected plate
output_image = orig.copy()
cv2.drawContours(output_image, [plate], -1, (0, 255, 0), 3)

# -------------------------------
# STEP 5: MASK & CROP
# -------------------------------
mask = np.zeros(gray.shape, np.uint8)
cv2.drawContours(mask, [plate], 0, 255, -1)

(x, y) = np.where(mask == 255)
(topx, topy) = (np.min(x), np.min(y))
(bottomx, bottomy) = (np.max(x), np.max(y))

cropped = gray[topx:bottomx+1, topy:bottomy+1]

# Threshold for characters
_, thresh = cv2.threshold(cropped, 0, 255,
                          cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# -------------------------------
# STEP 6: CHARACTER SEGMENTATION
# -------------------------------
cnts, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

char_regions = []

for c in cnts:
    x, y, w, h = cv2.boundingRect(c)

    aspect_ratio = w / float(h)

    # Improved filtering
    if 20 < h < 200 and 0.2 < aspect_ratio < 1.0:
        char = thresh[y:y+h, x:x+w]
        char = cv2.resize(char, (20, 20))
        char_regions.append((x, char))

# Sort left to right
char_regions = sorted(char_regions, key=lambda x: x[0])
characters = [c[1] for c in char_regions]

# -------------------------------
# STEP 7: LOAD DATASET (ZIP)
# -------------------------------
zip_path = "archive (1).zip"
extract_path = "dataset"

# Ensure the extract path exists
if not os.path.exists(extract_path):
    os.makedirs(extract_path)
    print(f"Created directory: {extract_path}")
else:
    print(f"Directory '{extract_path}' already exists.")

extracted_successfully = False
# Check if the directory is already populated
if os.listdir(extract_path):
    print(f"Directory '{extract_path}' is already populated. Skipping extraction.")
    extracted_successfully = True
else:
    print(f"Attempting to extract '{zip_path}' to '{extract_path}'...")
    try:
        # First attempt with zipfile module
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        if os.listdir(extract_path):
            print("Extraction successful using zipfile module.")
            extracted_successfully = True
        else:
            print("Zipfile module extraction failed to populate the directory.")
    except zipfile.BadZipFile:
        print(f"Error: '{zip_path}' is a bad zip file.")
    except FileNotFoundError:
        print(f"Error: Zip file '{zip_path}' not found.")
    except Exception as e:
        print(f"An unexpected error occurred during zipfile extraction: {e}")

    # Fallback to 'unzip' command if zipfile failed or didn't populate
    if not extracted_successfully or not os.listdir(extract_path):
        print("Falling back to 'unzip' command for extraction...")
        try:
            command = ['unzip', '-o', zip_path, '-d', extract_path]
            result = subprocess.run(command, capture_output=True, text=True)
            if result.returncode == 0 and os.listdir(extract_path):
                print("Extraction successful using 'unzip' command.")
                # print("Unzip stdout:\n", result.stdout) # Optional: print verbose output
                extracted_successfully = True
            else:
                print(f"Extraction with 'unzip' command failed. Return code: {result.returncode}")
                # print("Unzip stdout:\n", result.stdout)
                # print("Unzip stderr:\n", result.stderr)
        except FileNotFoundError:
            print("Error: 'unzip' command not found. Please ensure it's installed.")
        except Exception as e:
            print(f"An unexpected error occurred during 'unzip' command execution: {e}")

if not extracted_successfully or not os.listdir(extract_path):
    print("FATAL ERROR: Could not extract dataset. Please check 'archive (1).zip' or the extraction path.")
    # If extraction consistently fails, uncomment `exit()` to stop execution here.
    # exit()


data = []
labels = []

image_extensions = ('.png', '.jpg', '.jpeg', '.bmp')

if not os.listdir(extract_path):
    print(f"Warning: Dataset directory '{extract_path}' is empty. No images will be loaded.")
else:
    for label in os.listdir(extract_path):
        folder = os.path.join(extract_path, label)

        if os.path.isdir(folder):
            for img_name in os.listdir(folder):
                if not img_name.lower().endswith(image_extensions):
                    continue

                img_path = os.path.join(folder, img_name)
                img = cv2.imread(img_path, 0)

                if img is None:
                    print(f"Warning: Could not read image {img_path}. Skipping.")
                    continue

                img = cv2.resize(img, (20, 20))
                data.append(img.flatten())
                labels.append(label)

if len(data) == 0:
    print("FATAL ERROR: No images were loaded from the dataset directory. Cannot proceed with model training.")
    # This will lead to the ValueError if data is empty, but this message clarifies the issue.

data = np.array(data) / 255.0
labels = np.array(labels)

# Add train-test split for model evaluation
if len(data) > 0:
    X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
    print(f"Dataset split into {len(X_train)} training samples and {len(X_test)} testing samples.")
else:
    X_train, X_test, y_train, y_test = [], [], [], [] # Ensure they are defined even if data is empty
    print("No data available for train-test split.")

# -------------------------------
# STEP 8: TRAIN SVM
# -------------------------------
if len(X_train) == 0:
    print("FATAL ERROR: No training data available. Cannot train SVM model.")
    # This will prevent the sklearn ValueError by providing a custom message.
else:
    model = make_pipeline(StandardScaler(), svm.SVC(kernel='linear'))
    model.fit(X_train, y_train)
    print("SVM model trained successfully.")

    # Calculate and print accuracy on the test set
    if len(X_test) > 0:
        accuracy = model.score(X_test, y_test)
        print(f"Model accuracy on test set: {accuracy:.4f}")
    else:
        print("No test data available to calculate accuracy.")

# -------------------------------
# STEP 9: PREDICTION
# -------------------------------
if len(characters) == 0:
    print("No characters detected")
    exit()

features = np.array([c.flatten() for c in characters]) / 255.0

# Only attempt prediction if the model was trained (i.e., X_train was not empty)
if len(X_train) > 0:
    predictions = model.predict(features)
    plate_text = ''.join([str(p) for p in predictions])
    print("Detected Plate:", plate_text)
else:
    print("Skipping plate prediction as the model could not be trained due to missing data.")

# -------------------------------
# STEP 10: OUTPUT
# -------------------------------
print("\n--- PROCESS OUTPUTS ---")

cv2_imshow(orig)
print("Original Image")

cv2_imshow(gray)
print("Grayscale")

cv2_imshow(binary)
print("Binarized")

cv2_imshow(edges)
print("Edges")

cv2_imshow(output_image)
print("Detected Plate Region")

cv2_imshow(cropped)
print("Cropped Plate")

cv2_imshow(thresh)
print("Final Binary Plate")

print("\n--- CHARACTERS ---")
for i, char in enumerate(characters):
    print(f"Character {i+1}")
    cv2_imshow(char)
```
